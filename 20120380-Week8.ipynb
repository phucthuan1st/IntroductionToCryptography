```python
from math import sqrt

def isPerfectSquare(x):
    root = sqrt(x)
    return pow(int(root + 0.5), 2) == x

def factorize(n):
    b = 1
    
    while True:
        if b == n:
            return (None, None)
        
        a = n + pow(b,2)
        
        print(f"{n} + {b}^2 = {a}")
        
        if isPerfectSquare(a):
            root = int(sqrt(a))
            
            first_factor = root + b
            second_factor = root - b
            
            return (first_factor, second_factor)
        
        b = b + 1
```

```python
# a
print(factorize(53357))

# b
print(factorize(34571))

#c
print(factorize(25777))

#d 
print(factorize(64213))
```

Xét giới hạn: $H = \lim\limits_{X \rightarrow \infty}{\dfrac{L(X)}{{ln(X)}^\alpha}}$
$=\lim\limits_{X \rightarrow \infty}{\dfrac{e^{\sqrt{lnX \cdot lnlnX}}}{{ln(X)}^\alpha}}$

$=\lim\limits_{X \rightarrow \infty}{e^{ln{\dfrac{e^{\sqrt{lnX \cdot lnlnX}}}{X^\beta}}}}$

$=\lim\limits_{X \rightarrow \infty}{e^{\sqrt{lnX \cdot lnlnX} -ln{X^\beta}}} \rightarrow e^{-\infty}$

Dễ thấy khi $X \rightarrow \infty$ thì $H \rightarrow 0 < \infty$

Do đó ta nói: $L(X) = O(X^\beta)$

## 3.36

### a)
${a^{\frac{p-1}{2}}}^2 = a^{p-1} ≡ 1 \text{ (mod p)}$ 

Như vậy $ {a^{\frac{p-1}{2}}}^2$ là một nghiệm đồng dư $x^2 ≡ 1 \text{ (mod p)}$

Phương trình đồng dư trên có hai nghiệm là $x = 1$ và $x = -1$.

Như vậy $ {a^{\frac{p-1}{2}}}^2 ≡ ±1 \text{ (mod p)}$

### b)
Ta có ${a^{\frac{p-1}{2}}}^2 ≡ ±1 \text{ (mod p)}$ nên $a^2 ≡ 2^{p−1} \text{ (mod p)}$

Theo định lí Fermat nhỏ ta ta có:

$ a^{p-1} \cdot (p-1)! ≡ (p-1)! \text{ (mod p)}$

$\Rightarrow a^{p-1} = a^p-a ≡ 1 \text{ (mod p)}$

Như vậy ta được $ a^2=2^{p−1} ≡ 1 \text{ (mod p)}$
